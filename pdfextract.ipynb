```python
import pdfplumber

def show_char(start , end, pdf, char):
    charlist=[]
    for i in range(start, end):
        charlist.append(str(i) + ':' + pdf.chars[i][char])
    return charlist


def get_BboxPos(start, end, pdf):
    dic_char = {
        'x0':pdf.chars[start]['x0'],
        'top':pdf.chars[start]['top'],
        'x1':pdf.chars[end]['x1'],
        'bottom':pdf.chars[end]['bottom']
    }
    return dic_char

def get_text(pdf, dict_pos):
    crop = pdf.within_bbox((dict_pos['x0'],dict_pos['top'],dict_pos['x1'],dict_pos['bottom']))
    return crop.extract_text()

def search(aList , searchFor):
    res = []
    for value in aList:
        if value.find(searchFor) > 0 :
            res.append(value)
    return res

def runExtract(file):
    pdf_1 = pdfplumber.open(file)
    p0 = pdf_1.pages[0]

    print('字元總數:'+ str(len(p0.chars)))
 
    mylist = show_char(0, len(p0.chars), p0, 'text')
    
    mystr = ''
    
    while mystr != 'end':
        mystr = input('請輸入要找尋的文字(離開請輸入end):')
        if mystr == 'end':
            break
        
        s = ''
        e = ''
        while s != 'end' or e != 'end':
            print(search(mylist, mystr))
            s = input('文字開始序號(離開請輸入end):')
            if s == 'end':
                break
            e = input('文字結束序號(離開請輸入end):')
            if e == 'end':
                break

            box = get_BboxPos(int(s), int(e), p0)
            text = get_text(p0, box)
            print('-----------------------------------------------------------')
            print('解析文字:')
            print(text)
            print('')
            print('位置資訊:')
            print(box)
            print('-----------------------------------------------------------')


if __name__ == "__main__":
    file = 'd:/N2.pdf'
    runExtract(file)
    
    ans = input('要以位置搜尋文字嗎?(y/n)')
    if ans == 'y':
        pdf = pdfplumber.open(file)
        p0 = pdf.pages[0]
        

        pos = {
        'x0':None,
        'top':None,
        'x1':None,
        'bottom':None
        }
        gap = [0, 36.44, 36.48, 36.48, 36.44]
        multi = 1
        pos = {'x0':448.640, 'top':190.344, 'x1': 454.640, 'bottom':208.200}
        
        print('開始文字:' , get_text(p0, pos))
        for i in range(5):
            pos['top'] = pos['top']+gap[i]
            pos['bottom'] = pos['bottom']+gap[i]
            print(i, get_text(p0, pos), gap[i])

#         while get_text(p0, pos) == None:
#             pos['top'] = pos['top']+0.01
#             pos['bottom'] = pos['bottom']+0.01
#             gap = gap+0.01
        
        
#         for key in pos.keys():
#             pos[key] = input('請輸入' + key + '的數值:')
        
        
#         print('結束文字', get_text(p0, pos), 'at', gap)
        
```
